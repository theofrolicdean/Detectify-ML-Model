```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

%matplotlib inline
warnings.filterwarnings("ignore")
```

```python
AI_DATASET_PATH = "datasets/ai_generated_indonesia_combined.csv"
HUMAN_DATASET_PATH = "datasets/human_generated_indonesia_combined.csv"
```

```python
ai_df = pd.read_csv(AI_DATASET_PATH)
human_df = pd.read_csv(HUMAN_DATASET_PATH)
```

```python
ai_df = ai_df.drop(columns=["topic"])
```

```python
ai_df.head()
```

```python
ai_df = ai_df.rename(columns={"text": "message"})
```

```python
ai_df.head()
```

```python
human_df = human_df.drop(columns=["Unnamed: 0"])
```

```python
human_df.head()
```

```python
combined_df = pd.concat([ai_df, human_df], ignore_index=True)
```

```python
combined_df = combined_df.sample(frac=1, random_state=42).reset_index(drop=True)
```
