```python
from transformers import GPT2TokenizerFast, GPT2LMHeadModel
pretrained_name = "w11wo/indo-gpt2-small"
tokenizer = GPT2TokenizerFast.from_pretrained(pretrained_name)
tokenizer.model_max_length = 1024
model = GPT2LMHeadModel.from_pretrained(pretrained_name)
```

```python
total_sentences = 50000
batch_size = 5
```

```python
topics = [
    "Tentang teknologi: ",
    "Tentang pendidikan: ",
    "Tentang makanan: ",
    "Tentang olahraga: ",
    "Tentang lingkungan: ",
    "Tentang budaya Indonesia: ",
    "Tentang kesehatan: ",
    "Tentang musik: ",
    "Tentang perjalanan: ",
    "Tentang sains: "
]
```

```python
filename = "datasets/ai_generated_indonesia.csv"
import torch
import csv
import random
import time
from tqdm import tqdm
```

```python
with open(filename, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["message"])

    generated = 0
    # Progress bar
    with tqdm(total=total_sentences, desc="Generating Indonesian Sentences") as pbar:
        while generated < total_sentences:
            prompt = random.choice(topics)
            input_ids = tokenizer.encode(prompt, return_tensors='pt')

            # Generate sentences
            sample_outputs = model.generate(
                input_ids,
                pad_token_id=50256,
                do_sample=True,
                max_length=40,
                min_length=20,
                top_k=50,
                num_return_sequences=batch_size
            )

            for sample_output in sample_outputs:
                text = tokenizer.decode(sample_output.tolist(), skip_special_tokens=True)
                writer.writerow([text.strip()])
                generated += 1
                pbar.update(1)  # update progress bar
                if generated >= total_sentences:
                    break

            time.sleep(0.5)
```
