```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

%matplotlib inline
warnings.filterwarnings("ignore")
```

```python
INDONESIAN_NEWS_DATASET_PATH = "datasets/indonesian_news_dataset.csv"
SENTIMENT_DATASET_PATH = "datasets/sentiment.csv"
WIKIPEDIA_DATASET_PATH = "datasets/wikipedia_id-clean.csv"
```

```python
indonesian_news_df = pd.read_csv(INDONESIAN_NEWS_DATASET_PATH)
```

```python
indonesian_news_df.head()
```

```python
indonesian_news_df = indonesian_news_df[["content"]]
```

```python
indonesian_news_df.head()
```

```python
indonesian_news_df["content"] = (
    indonesian_news_df["content"]
    .str.replace(r"^.*?\s-\s", "", regex=True)
    .str.strip()
)
```

```python
indonesian_news_df.head()
```

```python
indonesian_news_df.tail()
```

```python
indonesian_news_df.shape
```

```python
sentiment_df = pd.read_csv(SENTIMENT_DATASET_PATH)
```

```python
sentiment_df.head()
```

```python
sentiment_df = sentiment_df[["kalimat"]]
```

```python
sentiment_df.shape
```

```python
sentiment_df.head()
```

```python
wikipedia_df = pd.read_csv(WIKIPEDIA_DATASET_PATH)
```

```python
wikipedia_df.head()
```

```python
wikipedia_df = wikipedia_df[["text"]]
```

```python
wikipedia_df = wikipedia_df.sample(frac=0.1, random_state=42).reset_index(drop=True)
```

```python
wikipedia_df.head()
```

```python
wikipedia_df.shape
```

```python
indonesian_news_df = indonesian_news_df.rename(columns={"content": "message"})
sentiment_df = sentiment_df.rename(columns={"kalimat": "message"})
wikipedia_df = wikipedia_df.rename(columns={"text": "message"})
```

```python
combined_df = pd.concat(
    [indonesian_news_df, sentiment_df, wikipedia_df],
    ignore_index=True
)
```

```python
combined_df["label"] = 0
```

```python
combined_df = combined_df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
combined_df.shape
```

```python
combined_df.head()
```

```python
combined_df.tail()
```

```python
combined_df.to_csv("datasets/human_generated_indonesia_combined.csv")
```
